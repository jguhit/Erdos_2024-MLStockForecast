Atricle Counts by Date

In this notebook we look at graphs of article count against date, and we calculate for each ticker, the total number of days where more than 5 articles were published about that ticker.

```python
import pandas as pd
import numpy as np
import datetime
from seaborn import set_style
import plotly as py
import matplotlib.pyplot as plt

set_style("whitegrid")
```

```python
finv = pd.read_csv('../data/data_finvader.csv')
```

```python
#finv.info()
```

Section 1: Technology

1.1 APPLE

```python
#APPLE

#restricting to Apple
Apl = finv.loc[finv["Ticker"] == 'AAPL']
#Counting the article number
Apl1 = Apl.groupby('Date')['Ticker'].count().reset_index()
Apl2 = Apl1.rename(columns={"Ticker": "Article_Count" })
#Adding the missing days
Apl2.index = pd.DatetimeIndex(Apl2['Date']).floor('D')
all_days = pd.date_range(start='2019-03-15', end='2024-03-15', freq='D')
Apl2 = Apl2.reindex(all_days)
#Adding 0's for missing article counts
Apl2['Article_Count'] = Apl2['Article_Count'].fillna(0)
```

```python
#Testing if the above code worked. (ALL GOOD)
#Apl2.head()
```

```python
#grpah of Apple article count over date

import plotly.express as px

fig = px.line(Apl2, x=Apl2.index, y="Article_Count", title="Apple")
fig.show()
```

```python
#number of days with more than 5 articles
Apl2.loc[Apl2["Article_Count"] >= 6].count()
```
