```python
import pandas as pd
import random
import os
from bs4 import BeautifulSoup
import requests
```

```python
current_dir = os.getcwd()
data_dir = os.path.join(current_dir, 'data')
csv_file = os.path.join(data_dir, 'news_data_2019_2024_sorted.csv')
df = pd.read_csv(csv_file, parse_dates=['Date & Time'])
```

```python
df.set_index('Date & Time', inplace=True)
df.sort_index(inplace=True)
```

```python
def scrape_engagement_metrics(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
    '''
    views_span = soup.find('span', class_='views-count')
    if views_span:
        views = views_span.text
        return int(views.replace(',', ''))
    else:
        return None
    '''
```

```python
first_url = df.iloc[0]['URL']
print(f"Scraping engagement metrics for: {first_url}")
```

```python
page_views = scrape_engagement_metrics(first_url)
print(f"Page Views: {page_views}")
```
